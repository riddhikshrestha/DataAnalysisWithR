```r
library(readxl)
library(ggplot2)
```

```r
data <- read_excel('data.xlsx')
```

```r
head(data)
```

```r
colnames(data)
```

```r
summary(data)
```

```r
# Setup factors
data$Player_name <- as.factor(data$Player_name)
data$Gender <- as.factor(data$Gender)
data$Age <- as.factor(data$Age)
data$Negotiation <- as.factor(data$Negotiation)
data$Collaboration <- as.factor(data$Collaboration)
data$Communication <- as.factor(data$Communication)
```

```r
#Player Negotiation
ggplot(data,aes(x=Negotiation))+geom_bar()
```

```r
ggplot(data,aes(x=Collaboration))+geom_bar()
```

```r
ggplot(data,aes(x=Communication))+geom_bar()
```

```r
#Negotiation in percentages
prop.table(table(data$Negotiation))
```

```r
# Add some customization for labels and theme bw color
ggplot(data,aes(x=Negotiation))+theme_bw()+geom_bar()+labs(y="Players Negotiation",title="Player's Negotiation Rates")
```

```r
# Negotiation by gender in data of players
ggplot(data,aes(x=Gender,fill=Negotiation))+theme_gray()+geom_bar()+labs(y="Players Count",title="Negotiation Rate by Gender")
```

```r
# Negotiation of players by their age
ggplot(data,aes(x=Age,fill=Negotiation))+theme_grey()+geom_bar()+labs(y="Players Count",title="Players Negotiation Rates by their Age")
```

```r
# Negotiation of players by their age
ggplot(data,aes(x=Player_name,fill=Negotiation))+theme_grey()+geom_bar()+labs(y="Players Count",title="Players Negotiation Rates by their Name")
```
